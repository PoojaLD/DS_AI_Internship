```python
import sqlite3

# Connect to existing database
conn = sqlite3.connect("internship.db")
cursor = conn.cursor()

# Create mentors table
cursor.execute("""
CREATE TABLE IF NOT EXISTS mentors (
    mentor_id INTEGER PRIMARY KEY,
    mentor_name TEXT,
    track TEXT
)
""")

# Insert mentor data
mentor_data = [
    (1, "Dr. Sharma", "Data Science"),
    (2, "Mr. John", "Web Dev"),
    (3, "Ms. Kavya", "UI/UX")
]

cursor.executemany("INSERT OR IGNORE INTO mentors VALUES (?, ?, ?)", mentor_data)

conn.commit()
```
