```python
import pandas as pd
import numpy as np

# Example dataset (replace with your real data later)
data = {
    "Salary": [25000, 30000, 32000, 35000, 40000, 45000, 50000, 120000]  # last value is outlier
}

df = pd.DataFrame(data)

# -----------------------------
# 1️⃣ Mean and Standard Deviation
# -----------------------------
mean_val = df["Salary"].mean()
std_val = df["Salary"].std()

print("Mean:", mean_val)
print("Standard Deviation:", std_val)

# -----------------------------
# 2️⃣ Calculate Z-score
# -----------------------------
df["z_score"] = (df["Salary"] - mean_val) / std_val

print("\nDataset with Z-scores:")
print(df)

# -----------------------------
# 3️⃣ Identify Outliers |Z| > 3
# -----------------------------
outliers = df[np.abs(df["z_score"]) > 3]

print("\nOutliers detected:")
print(outliers)
```
