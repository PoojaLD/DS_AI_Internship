```python
import numpy as np

# Create a 5x3 array of random scores (50 to 100)
scores = np.random.randint(50, 101, size=(5, 3))

# Column-wise mean
subject_means = np.mean(scores, axis=0)

# Broadcasting subtraction (normalization)
centered_scores = scores - subject_means

print("Original Scores:\n", scores)
print("\nSubject Means:\n", subject_means)
print("\nCentered Scores:\n", centered_scores)
```
