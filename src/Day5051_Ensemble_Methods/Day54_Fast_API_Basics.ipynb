Work Summary

Today, I transitioned from model building to Web API Development using the Flask framework. I learned that an API (Application Programming Interface) acts as the essential bridge between a machine learning model and the end-user. I successfully built a lightweight web server, defined custom Routes, and implemented the two primary HTTP methods: GET (for retrieving data) and POST (for sending data to the server). I also mastered JSON handling, ensuring my API can consume and return data in the universal format required by modern web applications. Finally, I practiced testing my endpoints using Postman and cURL to verify that my backend logic was functioning correctly before structuring the project for a professional GitHub repository.


Hours worked 4.5

Learnings / Outcomes
The API Bridge: Realized that an API is the "waiter" of the software world—taking a request from the client (frontend), bringing it to the kitchen (ML model/Logic), and returning the result (Prediction).
Routing & HTTP Verbs: Understood the distinction between GET (safe, idempotent requests like fetching a model's version) and POST (sending a feature vector to the model for a prediction).
JSON Serialization: Mastered the use of request.get_json() and Python dictionaries to exchange data. I learned that because JSON is language-agnostic, my Python model can now "talk" to a JavaScript frontend or a Java mobile app.
The Development Loop: Established a workflow using app.run(debug=True), which automatically restarts the server when code changes, significantly speeding up the debugging process.

Blockers / Risks
The "Browser POST" Trap: I initially tried to test my POST routes by typing the URL into my browser, only to receive a "Method Not Allowed" error. I learned that browsers default to GET requests and that tools like Postman or cURL are mandatory for testing data submissions.
Port Conflicts: I encountered an "Address already in use" error. I had to learn how to identify background processes or simply change the port using app.run(port=5001).
Silent JSON Failures: I struggled when my API returned a 400 Bad Request. I discovered this was due to sending a malformed JSON body in Postman. I now know to always validate my JSON syntax (commas, quotes, etc.) before hitting "Send."

Skills Used

Backend Web Development, Flask Framework, RESTful API Design, HTTP Methods (GET/POST), JSON Data Handling, API Testing (Postman/cURL), Project Structuring.
