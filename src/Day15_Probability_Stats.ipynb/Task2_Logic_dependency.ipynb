```python
# -----------------------------
# 1️⃣ Independent Events
# Coin = Heads, Dice = 6
# -----------------------------
P_heads = 1/2
P_six = 1/6

P_independent = P_heads * P_six

print("Probability of Heads AND rolling 6:", P_independent)


# -----------------------------
# 2️⃣ Dependent Events
# Picking 2 red marbles without replacement
# -----------------------------
red_marbles = 5
total_marbles = 10

# First pick red
P_first_red = red_marbles / total_marbles

# After removing one red marble
P_second_red = (red_marbles - 1) / (total_marbles - 1)

P_dependent = P_first_red * P_second_red

print("Probability of picking 2 Red marbles:", P_dependent)
```
