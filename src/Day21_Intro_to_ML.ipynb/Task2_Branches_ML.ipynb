    1️⃣ Supervised Learning

Supervised Learning is a type of Machine Learning where the data contains both inputs and correct outputs (labels). The model learns a mapping between input features and the known answers so it can predict outcomes for new data. It is commonly used for classification and regression problems.

    2️⃣ Unsupervised Learning

Unsupervised Learning is used when the data does not have labels. The model tries to discover hidden patterns, structures, or clusters within the data. It is mainly used for grouping similar data points or reducing dimensionality.

    3️⃣ Reinforcement Learning

Reinforcement Learning is a type of learning where an agent interacts with an environment and learns by receiving rewards or penalties based on its actions. The goal is to maximize long-term rewards through trial and error.
